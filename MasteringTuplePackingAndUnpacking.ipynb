```python
"""
Objective: The goal of this assignment is to deepen your understanding of tuple packing and unpacking in Python.

Task 1: Customer Order Processing 
Refine your skills in tuple unpacking by managing customer orders.

Problem Statement: You are given a list of tuples, each representing a customer's order. Each tuple contains the customer's name, the product ordered, and the quantity. Your task is to unpack each tuple and print the order details in a user-friendly format.

Sample Order Data:

orders = [
    ("Alice", "Laptop", 1),
    ("Bob", "Camera", 2),
    # More orders...
]
- Write a function to iterate over the list of orders. 
- Unpack each tuple in the list and format the details for display.
"""
```

```python
# order = (name, item, quantity)

def add_order(orders):
    try:
        name = input("\nEnter the customer's name: ")
        if name == "":
            print("You must enter a customer's name to continue.")
        else:
            product = input("\nEnter the product ordered: ")
            if product == "":
                print("You must enter a product to continue.")
            else:
                quantity = int(input("\nEnter the quantity ordered: "))
                if quantity == "":
                    print("You must enter a quantity to continue.")
                else:
                    new_order = (name, product, quantity)
                    orders.append(new_order)
                    print(f"\n* New order added to be processed: *\n\t~ Customer's Name: {new_order[0].capitalize()}\n\t~ Product: {new_order[1].capitalize()}\n\t~ Quantity: {new_order[-1]}")
    except ValueError:
        print("\nInvalid input. Quantity must be a number.")

def display_orders(orders):
    for order in orders:
        print(f"\n\t~ Customer's Name: {order[0].capitalize()}\n\t~ Product: {order[1].capitalize()}\n\t~ Quantity: {order[-1]}")

def order_processing():
    orders = [
    ("Alice", "Laptop", 1),
    ("Bob", "Camera", 2)
]
    def menu():
        # Print menu
        print("\n~ Order Processing System ~")
        print("1. Add Order")
        print("2. Display Orders")
        print("3. Exit")

        # Ask user what they want to do
        user_input = input("Enter your choice: ")
        try:
            choice = int(user_input)
        except ValueError:
            print("\nThat is not a valid choice. Please enter the digit that corresponds with your selection.")
        except TypeError:
            print("\nAnd unexpected type error occurred. Please ensure you enter the digit that corresponds with your selection.")
        else:
            if choice == 1:
                add_order(orders)
                menu()
            elif choice == 2:
                print("\n* All Customer Orders: *")
                display_orders(orders)
                menu()
            elif choice == 3:
                print("\nThank you for using Order Processing System!")

    try:
        menu()
    finally:
        print("Exiting program...")

order_processing()
```
